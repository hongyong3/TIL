```python
import http.client, json

conn = http.client.HTTPSConnection("api.themoviedb.org")

payload = "{}"
dataObject = []
for i in range(10):
    conn.request("GET", "/3/movie/now_playing?page="+f"{i+1}"+"&language=ko-kr&api_key=f55decdd8b4533ff751d5e5df47ac372", payload)
    # print("/3/movie/now_playing?page=" + f"{i+1}" + "&language=ko-kr&api_key=f55decdd8b4533ff751d5e5df47ac372")
    res = conn.getresponse()
    data = res.read()
    dataObject += json.loads(data)['results']
print(dataObject)

with open('movies.json', 'w', encoding='utf-8') as f:
    json.dump(dataObject, f, ensure_ascii=False, indent=4)
#     print(data.decode("utf-8"))
    
```
